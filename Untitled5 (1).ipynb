# VISUALIZING THE DATASET

### VISUALIZING SALES OVER TIME

```python
# Convert ORDERDATE to datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], format='%m/%d/%Y %H:%M')

# Plot Sales Over Time
plt.figure(figsize=(14, 7))
sales_over_time = df.groupby('ORDERDATE').sum()['SALES']
sales_over_time.plot()
plt.title('Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

#### This line plot shows how sales have changed over time. It provides a time-series view to identify trends and patterns.

```python
# Plot Sales by Product Line
plt.figure(figsize=(14, 7))
sns.boxplot(x='PRODUCTLINE', y='SALES', data=df)
plt.title('Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

#### This box plot shows the distribution of sales for each product line. It helps identify which product lines generate the most revenue and the variability within each category.

```python
# Plot Sales Distribution by Deal Size
plt.figure(figsize=(14, 7))
sns.boxplot(x='DEALSIZE', y='SALES', data=df)
plt.title('Sales Distribution by Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Sales')
plt.show()
```

#### This box plot shows the distribution of sales for each product line. It helps identify which product lines generate the most revenue and the variability within each category.


```python
plt.figure(figsize=(14, 7))
country_sales = df.groupby('COUNTRY').sum(numeric_only=True)['SALES'].sort_values(ascending=False)
country_sales.plot(kind='bar')
plt.title('Geographic Distribution of Sales')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.show()
```

#### This bar plot shows the total sales per country. It helps identify the top-performing regions.

```python
import folium
from geopy.geocoders import Nominatim

country_sales = df.groupby('COUNTRY').sum(numeric_only=True)['SALES'].reset_index()

# Geocode country names to get latitude and longitude
geolocator = Nominatim(user_agent="geoapiExercises")

def get_lat_lon(country):
    try:
        location = geolocator.geocode(country)
        return location.latitude, location.longitude
    except:
        return None, None

country_sales['latitude'], country_sales['longitude'] = zip(*country_sales['COUNTRY'].apply(get_lat_lon))

# Remove countries with missing geocode data
country_sales = country_sales.dropna(subset=['latitude', 'longitude'])

# Create the map
m = folium.Map(location=[20, 0], zoom_start=2)

# Add sales data to the map
for _, row in country_sales.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=10,
        popup=f"{row['COUNTRY']}: ${row['SALES']:,.2f}",
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(m)

# Save the map to an HTML file
m.save('sales_map.html')

# Display the map
m
```
